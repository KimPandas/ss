#### 12.1.4. groupby í•¨ìˆ˜ë¡œ í–‰ ê°„ì˜ ì—°ì‚°í•˜ê¸°(shift ì™¸)

- `groupby` + `shift`: ê·¸ë£¹ ë‚´ì—ì„œ ë°ì´í„° ì´ë™

- `groupby` + `diff`: ê·¸ë£¹ ë‚´ì—ì„œ í–‰ê°„ì˜ ì°¨ì´ ê³„ì‚°

- `groupby` + `pct_change`: ê·¸ë£¹ ë‚´ì—ì„œ í–‰ê°„ì˜ ë³€ë™ë¥  ê³„ì‚°


````python
# ì½”ë“œ 12-14. ê·¸ë£¹ì„ ë‚˜ëˆ„ì–´ í–‰ ê°„ì˜ ì—°ì‚° ì‹¤ìŠµ ì˜ˆì œ ì½”ë“œ
import pandas as pd
pd.options.display.max_rows = 8 # 8í–‰ê¹Œì§€ë§Œ ì¶œë ¥
data = [['2025-01-03', 'íŒë‹¤ì „ì', 10000],
        ['2025-01-03', 'ì„±ì‹¬ë‹¹', 2000],
        ['2025-01-04', 'ì„±ì‹¬ë‹¹', 1600],
        ['2025-01-04', 'íŒë‹¤ì „ì', 12000],
        ['2025-01-05', 'íŒë‹¤ì „ì', 15000],
        ['2025-01-05', 'ì„±ì‹¬ë‹¹', 2000],
        ['2025-01-06', 'íŒë‹¤ì „ì', 13500],
        ['2025-01-06', 'ì„±ì‹¬ë‹¹', 2400]]

df = pd.DataFrame(data, columns=['ë‚ ì§œ', 'ì¢…ëª©', 'ì¢…ê°€'])
df
````

````python
# ì½”ë“œ 12-15. ê·¸ë£¹ì„ ë‚˜ëˆ„ì–´ ì „ì¼ì¢…ê°€ ì—´ ìƒì„±í•˜ê¸°
df['ì „ì¼ì¢…ê°€'] = df.groupby('ì¢…ëª©')['ì¢…ê°€'].shift()
df
````

````python
# ì½”ë“œ 12-16. ê·¸ë£¹ì„ ë‚˜ëˆ„ì–´ ì „ì¼ ëŒ€ë¹„ ë“±ë½ ì—´ ìƒì„±í•˜ê¸°
df['ë“±ë½'] = df.groupby('ì¢…ëª©')['ì¢…ê°€'].diff()
df
````

````python
# ì½”ë“œ 12-17. ê·¸ë£¹ì„ ë‚˜ëˆ„ì–´ ì „ì¼ ëŒ€ë¹„ ë“±ë½ë¥  ì—´ ìƒì„±í•˜ê¸°
df['ë“±ë½ë¥ '] = df.groupby('ì¢…ëª©')['ì¢…ê°€'].pct_change()
df
````

#### 12.1.5. groupby í•¨ìˆ˜ë¡œ ì „í›„ë°© ê°’ìœ¼ë¡œ ê²°ì¸¡ê°’ ëŒ€ì²´í•˜ê¸°(ffill, bfill)

- `groupby` + `ffill`: ê°™ì€ ê·¸ë£¹ì˜ ì „ë°©ê°’ìœ¼ë¡œ ê²°ì¸¡ê°’ ëŒ€ì²´

- `groupby` + `bfill`: ê°™ì€ ê·¸ë£¹ì˜ í›„ë°©ê°’ìœ¼ë¡œ ê²°ì¸¡ê°’ ëŒ€ì²´


````python
# ì½”ë“œ 12-18. ê·¸ë£¹ì„ ë‚˜ëˆ„ì–´ ê²°ì¸¡ê°’ ëŒ€ì²´í•˜ê¸° ì‹¤ìŠµ ì˜ˆì œ ì½”ë“œ
import pandas as pd
data1 = [['2025-01-03', 'íŒë‹¤ì „ì', 10000, float('nan')],
         ['2025-01-03', 'ì„±ì‹¬ë‹¹', 2000, float('nan')],
         ['2025-01-04', 'ì„±ì‹¬ë‹¹', 1600, 1],
         ['2025-01-04', 'íŒë‹¤ì „ì', 12000, float('nan')],
         ['2025-01-05', 'íŒë‹¤ì „ì', 15000, 2],
         ['2025-01-05', 'ì„±ì‹¬ë‹¹', 2000, float('nan')],
         ['2025-01-06', 'íŒë‹¤ì „ì', 13500, float('nan')],
         ['2025-01-06', 'ì„±ì‹¬ë‹¹', 2400, float('nan')]]

df1 = pd.DataFrame(data1, columns=['ë‚ ì§œ', 'ì¢…ëª©', 'ì¢…ê°€', 'êµ¬ë§¤'])
df1
````

#### 12.1.6. groupby í•¨ìˆ˜ë¡œ ëˆ„ì  í•© êµ¬í•˜ê¸°(cumsum)

- `groupby` + `cumsum`: ê·¸ë£¹ ë‚´ì—ì„œ ëˆ„ì í•© ê³„ì‚°

````python
# ì½”ë“œ 12-23. ê·¸ë£¹ì„ ë‚˜ëˆ„ì–´ ëˆ„ì  í•© êµ¬í•˜ê¸° ì‹¤ìŠµ ì˜ˆì œ ì½”ë“œ
import pandas as pd
data = [['2021-01-01', 'ê¹€íŒë‹¤', 10000], ['2021-01-01', 'ê°•ìŠ¹ì£¼', 2000],
        ['2021-01-02', 'ê¹€íŒë‹¤', 20000], ['2021-01-02', 'ê°•ìŠ¹ì£¼', 5000],
        ['2021-01-03', 'ê°•ìŠ¹ì£¼', 8000], ['2021-01-03', 'ê¹€íŒë‹¤', 5000]]
df = pd.DataFrame(data, columns=['ë‚ ì§œ', 'ì´ë¦„', 'ì…ê¸ˆì•¡'])
df
````

````python
# ì½”ë“œ 12-24. dfì—ì„œ ê° ì‚¬ëŒì˜ ëˆ„ì  ì…ê¸ˆì•¡ êµ¬í•˜ê¸°
df['ëˆ„ì ì…ê¸ˆ'] = df.groupby('ì´ë¦„')['ì…ê¸ˆì•¡'].cumsum()
df
````

````python
# ì½”ë“œ 12-25. cumsum í•¨ìˆ˜ë¡œ ê·¸ë£¨í¼ ìƒì„± ì‹¤ìŠµ ì˜ˆì œ ì½”ë“œ
pd.options.display.max_rows = 10 # 10í–‰ê¹Œì§€ë§Œ ì¶œë ¥
data1 = [['07:35', 'A', 'Log_in', float('nan')],
         ['07:36', 'A', 'Buy', 2000.0],
         ['07:37', 'A', 'Log_out', float('nan')],
         ['10:30', 'B', 'Log_in', float('nan')],
         ['11:20', 'B', 'Buy', 4000.0],
         ['11:32', 'A', 'Log_in', float('nan')],
         ['11:34', 'B', 'Buy', 3000],
         ['11:35', 'B', 'Log_out', float('nan')],
         ['11:36', 'A', 'Buy', 5000.0],
         ['11:37', 'A', 'Log_out', float('nan')]]
df1 = pd.DataFrame(data1, columns=['ì‹œê°„', 'íšŒì›ì½”ë“œ', 'ë¡œê·¸', 'êµ¬ë§¤ê¸ˆì•¡'])
df1
````

````python
# ì½”ë“œ 12-26. ê° í–‰ì´ ë¡œê·¸ì¸ ë°ì´í„°ì¸ì§€ ë¶ˆ(bool)ë¡œ ë°˜í™˜í•˜ëŠ” ë¡œê·¸ì¸ ì—´ ìƒì„±
df1['ë¡œê·¸ì¸'] = df1['ë¡œê·¸'] == 'Log_in'
df1
````

````python
# ì½”ë“œ 12-27. íšŒì›ë§ˆë‹¤ ë¡œê·¸ì¸ì„ ê¸°ì¤€ìœ¼ë¡œ ê·¸ë£¹ ë‚˜ëˆ„ê¸°
df1['ê·¸ë£¹'] = df1.groupby('íšŒì›ì½”ë“œ')['ë¡œê·¸ì¸'].cumsum()
df1
````

````python
# ì½”ë“œ 12-28. íšŒì›ì½”ë“œ ì—´(index)ê³¼ ê·¸ë£¹ ì—´(columns)ë¡œ ë‚˜ëˆ„ì–´ êµ¬ë§¤ê¸ˆì•¡ì˜ í•© ì§‘ê³„
df1.pivot_table('êµ¬ë§¤ê¸ˆì•¡', index='íšŒì›ì½”ë“œ', columns='ê·¸ë£¹', aggfunc='sum')
````

#### 12.1.7. groupby í•¨ìˆ˜ë¡œ ìˆœë²ˆ êµ¬í•˜ê¸°(cumcount)

- `groupby` + `cumcount`: ê·¸ë£¹ ë‚´ì—ì„œ ìˆœë²ˆ êµ¬í•˜ê¸°

- `cumcount`ëŠ” ë°˜ë“œì‹œ ê·¸ë£¹ì„ ë‚˜ëˆ„ëŠ” í•¨ìˆ˜ì™€ í•¨ê»˜ ì‚¬ìš©


````python
# ì½”ë“œ 12-29. cumcount í•¨ìˆ˜ ì‹¤ìŠµ ì˜ˆì œ ì½”ë“œ
import pandas as pd
data = {'ì œí’ˆ': ['A', 'B', 'B', 'A', 'C', 'A'],
        'íŒë§¤ëŸ‰': [10, 20, 30, 40, 50, 60]}
df = pd.DataFrame(data)
df
````

````python
# ì½”ë“œ 12-30. ì œí’ˆìœ¼ë¡œ ê·¸ë£¹ì„ ë‚˜ëˆ„ì–´ ìˆœë²ˆ ìƒì„±í•˜ê¸°
df['ìˆœë²ˆ'] = df.groupby('ì œí’ˆ').cumcount()
df
````

````python
# ì½”ë“œ 12-31. ì œí’ˆ(index)ê³¼ ìˆœë²ˆ(columns)ë¡œ ê·¸ë£¹ì„ ë‚˜ëˆ„ì–´ íŒë§¤ëŸ‰ì„ í”¼ë²—
df.pivot(index='ì œí’ˆ', columns='ìˆœë²ˆ', values='íŒë§¤ëŸ‰')
````

````python
# ì½”ë“œ 12-32. ê·¸ë£¹ì„ ë‚˜ëˆ„ì§€ ì•Šì„ ë•Œ ì „ì²´ì˜ ìˆœë²ˆì„ ë§¤ê¸°ëŠ” ë°©ë²•
import numpy as np
df['ì „ì²´ìˆœë²ˆ'] = range(0, len(df)) # ê¸°ë³¸ íŒŒì´ì¬ì˜ ë°©ë²•
df['ì „ì²´ìˆœë²ˆ'] = np.arange(0, len(df)) # ë„˜íŒŒì´ì˜ ë°©ë²•
df['ì „ì²´ìˆœë²ˆ'] = pd.RangeIndex(0, len(df)) # íŒë‹¤ìŠ¤ì˜ ë°©ë²•
df
````

### ì—‘ì…€ ì˜ˆì œ 17. ì£¼ì‹ ohlcv ë°ì´í„°ë¡œ groupby í•¨ìˆ˜ë¥¼ ì‚¬ìš©í•´ ë‹¤ì–‘í•œ ì—´ ê°€ê³µí•˜ê¸°
