## 1. 시각화 기본

### 1.1. 시각화의 두 가지 접근 방식



데이터 시각화 라이브러리는 크게 두 가지 철학으로 나뉜다.

- 명령형(Imperative) 시각화: "그래프의 선은 빨간색으로 하고, 축은 이렇게 그려라"와 같이 모든 세부 과정을 일일이 지시하는 방식이다. 자유도가 높지만 코드가 길어진다. (예: Matplotlib)

- 선언형(Declarative) 시각화: "이 데이터를 사용해 막대 그래프를 그려라"와 같이 핵심 의도만 선언하는 방식이다. 빠르고 간결하지만 세부 수정이 어렵다. (예: Pandas plot, Seaborn)

## 2. Matplotlib 라이브러리 소개

### 2.1. Matplotlib 라이브러리란?

- Matplotlib은 파이썬 데이터 시각화의 근간이 되는 표준 라이브러리다.

- 시각화의 엔진: 대부분의 파이썬 시각화 도구는 내부적으로 Matplotlib의 기능을 사용한다.판다스의 `plot`, Seaborn 라이브러리 역시 마찬가지이다.

- 객체 지향 구조: 전체 도화지(Figure)와 개별 그래프 영역(Axes)을 조립하는 방식을 사용한다.

- 공식 문서: https://matplotlib.org/stable/

### 2.2. Matplotlib의 단점과 연계의 필요성

- Matplotlib의 단점: 데이터를 리스트나 배열로 매번 변환해야 하고, 단순한 그래프 하나를 그릴 때도 많은 코드가 필요하다.

- 연계의 이유: Pandas의 plot 함수는 데이터 프레임에서 바로 그래프를 그릴 수 있어 편리하다. Seaborn 라이브러리도 설정된 그래프는 손쉽게 그릴 수 있다. 따라서 **Pandas의 plot이나 Seaborn 라이브러리로 데이터의 뼈대를 빠르게 만들고, Matplotlib으로 디테일을 완성**하는 전략을 취한다.

## 3. Pandas plot 함수

### 3.1. plot 함수 소개

- `plot` 함수는 **Matplotlib 기반의 래퍼 함수**

- 복잡한 설정 없이 **데이터 전처리 → 시각화**를 한 줄로 연결할 수 있는 것이 가장 큰 장점  
- 즉, “빠르고 간단하게 그릴 수 있는 Matplotlib” 역할을 함


### 3.2. plot 함수의 장점
1. 자동 레이블링

- 데이터 프레임의 구조를 시각화 요소로 자동 매핑한다.

- 인덱스(Index) → X축: 데이터 프레임의 인덱스가 자동으로 그래프의 가로축 레이블이 된다.

- 컬럼명(Column) → 범례(Legend): 데이터 프레임의 각 컬럼은 서로 다른 색상의 선이나 막대로 표현되며, 컬럼명이 범례로 자동 지정된다.

- 장점: plt.xlabel()이나 plt.legend()를 일일이 설정하지 않아도 데이터 프레임 형태만 잘 잡혀 있다면 즉시 완성도 있는 그래프가 나온다.

2. 연속 메서드 (chain method)

- 데이터 처리 과정과 시각화를 끊김 없이 한 문장으로 연결할 수 있다.

  -  특징: 중간 결과를 별도의 변수에 저장할 필요가 없어 코드가 간결해진다.

  - 예시: df.groupby('city')['salary'].mean().plot(kind='bar')

3. 데이터 프레임 조작을 통한 손쉬운 그래프 변환

- 그래프 전용 함수를 배우는 대신, 익숙한 Pandas 함수로 데이터 프레임을 가공하여 시각적 디테일을 제어한다.

  - `rename_axis` 활용: 축의 이름(label)을 바꿀 때 데이터 프레임의 인덱스명을 변경하는 것만으로 그래프 라벨을 즉시 수정할 수 있다.

  - `set_index` 활용: 어떤 컬럼을 기준축(X축)으로 삼을지 데이터 프레임 수준에서 결정한다.

  - `sort_values` 활용: 데이터를 정렬하면 그래프의 막대나 선의 순서가 자동으로 정렬되어 가독성이 높아진다.

### 3.3. plot 함수로 다양한 그래프 그리기

**line, bar, barh, area, pie**는 데이터 프레임의 형태를 직관적으로 반영하므로 주로 `plot` 함수를 이용해 빠르게 생성한다.

```python
# 코드 1. 예제 데이터 로드
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

pd.options.display.max_rows = 6  # 6행까지만 출력

# 실습용 데이터 로드

cols = ['time', 'day', 'smoker', 'sex']
df_tips = sns.load_dataset('tips').astype({k: 'str' for k in cols})
df_tips
```

#### 3.3.1. line 그래프

시간의 흐름이나 순서에 따른 데이터의 변화 추세를 파악할 때 사용한다.

- 설정: `kind='line'` (기본값이므로 생략 가능)

- 특징: 인덱스가 X축이 되고, 각 컬럼이 개별 선으로 표시된다.

```python
# 코드 2. line 그래프 생성
# df_tips의 total_bill 열을 line 그래프로 그리기
df_tips['total_bill'].plot(kind='line')
```

#### 3.3.2. bar 그래프
범주형 데이터 간의 수치를 비교할 때 사용한다.

- 설정: `kind='bar'`

- 특징: 이미 합계나 평균으로 집계된 데이터를 시각화할 때 가장 효율적이다.

```python
# 코드3. 요일별 평균 팁 비교(bar)
df_tips.groupby('day')['tip'].mean().plot(kind='bar')
```

#### 3.3.3. barh 그래프

barh는 수평 막대 그래프를 생성한다.

- 설정: `kind='barh'`

- 특징: 이미 합계나 평균으로 집계된 데이터를 시각화할 때 가장 효율적이다.

```python
# 코드 4. 요일별 평균 팁 비교(barh)
df_tips.groupby('day')['tip'].mean().plot(kind='barh')
```

#### 3.3.4. area 그래프

선 그래프를 기반으로 하되, 축과 선 사이의 영역을 색상으로 채운다. 시간 흐름에 따른 누적 합계를 보여줄 때 유리하다.

- 설정: `kind='area'`

```python
# 코드 5. 요일별 팁과 전체 금액의 누적 합계 (상위 10개 데이터만)
df_tips[['total_bill', 'tip']].head(10).plot(kind='area')
```

#### 3.3.5. pie 그래프

전체에 대한 각 부분의 비율을 한눈에 보여준다.

- 설정: `kind='pie'`

- 특징: 데이터 프레임의 특정 컬럼을 지정해야 하며, `autopct`를 통해 백분율을 표시하는 것이 좋다.

```python
# 코드 6. 요일별 방문 빈도 비율 확인
df_tips['day'].value_counts().plot(kind='pie', autopct='%.1f%%', startangle=90)
```

## 4. Seaborn 라이브러리

Seaborn은 Matplotlib을 기반으로 하지만, 통계 분석에 특화된 고급 인터페이스를 제공한다. 단순한 시각화를 넘어 데이터의 분포, 관계, 통계적 추정을 파악하는 데 최적화되어 있으며, 특히 매개변수 `hue`를 통해 범주별 비교를 한 줄의 코드로 해결할 수 있다.

```python
# 코드 7. 예제 데이터 로드
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

pd.options.display.max_rows = 6  # 6행까지만 출력

# 실습용 데이터 로드

cols = ['time', 'day', 'smoker', 'sex']
df_tips = sns.load_dataset('tips').astype({k: 'str' for k in cols})
df_iris = sns.load_dataset('iris')
```

### 4.1. 기본 시각화

- 기본 시각화 중 **histogram**과 **scatter plot**은 seaborn이 시각적 품질과 표현력에서 더 뛰어남  
    
- 스타일, 색상 팔레트, 밀도 표현 등이 기본값으로 잘 갖춰져 있어 같은 코드라도 더 깔끔하고 해석하기 쉬운 그래프가 만들어짐

#### 4.1.1. 히스토그램

- Seaborn에서는 `histplot` 을 사용한다.

- 수치 데이터의 빈도 분포를 확인한다.

- `kde=True` 옵션을 통해 부드러운 밀도 곡선을 추가할 수 있다.

```python
# 코드 8. 전체 식사 금액의 분포와 밀도 곡선 확인
sns.histplot(data=df_tips, x='total_bill', kde=True, bins=20, color='skyblue')
```

#### 4.1.2. 산점도 그래프(scatter plot)
- `scatterplot` 함수를 사용한다.
- 두 변수 간의 상관관계를 파악한다.
- `hue` 파라미터로 그룹을 나누는 것이 Seaborn 산점도의 핵심이다.

```python
# 코드 9. 식사 금액과 팁의 관계를 요일별 색상으로 구분
sns.scatterplot(
    data=df_tips, x='total_bill', y='tip', hue='day', style='time', s=100
)
```

### 4.2. 고급 시각화 (분포 및 다변량 분석)

#### 4.2.1. 박스 플롯 (box plot)

- Seaborn에서는 `boxplot` 함수로 생성한다.

- 데이터의 분포와 이상치를 한눈에 파악하기 위해 **5수치 요약**(Five-Number Summary)을 시각화한 그래프다.

<br>

---

**박스 플롯의 구성 요소**

<img src=https://i.postimg.cc/x8ty4wWR/image.png, width=600>

- 중앙값 (Median): 박스 내부의 선으로, 데이터를 크기 순으로 나열했을 때 가운데 위치한 값이다.

- 1사분위수 (Q1) / 3사분위수 (Q3): 데이터의 하위 25% 지점과 75% 지점을 나타내며, 이 차이를 **IQR**(Interquartile Range)이라 부른다.

- 수염 (Whiskers): 박스로부터 뻗어 나간 선으로, 보통 IQR의 1.5배 범위 내에 있는 데이터의 최솟값과 최댓값을 연결한다.

- 이상치 (Outliers): 수염의 범위를 벗어나는 데이터로, 보통 점이나 작은 원으로 별도 표기한다.



```python
# 코드 10. 요일별 팁 분포를 흡연 여부에 따라 비교
sns.boxplot(data=df_tips, x='day', y='tip', hue='smoker')
```

#### 4.2.2. 바이올린 플롯 (violin plot)

- box plot에 데이터 밀도(KDE)를 결합한 형태다. 데이터가 어디에 많이 몰려 있는지 곡선의 너비로 보여준다.

- `viloinplot` 함수로 생성한다.

```python
# 코드 11. 요일별 식사 금액 분포를 성별로 나누어 확인 (split=True로 결합)
sns.violinplot(data=df_tips, x='day', y='total_bill', hue='sex', split=True)
```

#### 4.2.3. KDE 그래프

- 히스토그램의 막대를 제거하고 확률 밀도만 부드러운 선으로 표현한다.

- `kdeplot`으로 생성한다.

```python
# 코드 12. 시간대별(점심/저녁) 식사 금액의 밀도 비교
sns.kdeplot(data=df_tips, x='total_bill', hue='time', fill=True, alpha=0.5)
```

#### 4.2.4. 조인트 플롯(joint plot)

- 두 수치형 변수의 산점도와 각 변수의 히스토그램을 결합하여 관계와 분포를 동시에 보여준다.

- `jointplot` 함수로 생성한다.

```python
# 코드 13. 식사 금액과 팁의 관계 및 각각의 분포를 한 번에 확인
sns.jointplot(data=df_tips, x='total_bill', y='tip', kind='reg')
```

#### 4.2.5. 페어 플롯(pair plot)

- 데이터 프레임 내의 모든 수치형 변수 간의 관계를 그리드 형태로 시각화한다.

- 데이터 탐색 초기에 필수적인 함수다.

- `pairplot` 함수로 수행한다.

```python
# 코드 14. 붓꽃의 품종별 모든 변수 간 상관관계 확인
sns.pairplot(data=df_iris, hue='species')
```

#### 4.2.6. 선형 회귀 그래프 (Regression Plot)

- 산점도 위에 회귀선과 신뢰구간을 그려 두 변수 간의 선형적 추세를 통계적으로 보여준다.

- `lmplot` 함수로 수행한다.

```python
# 코드 15. 흡연 여부에 따른 전체 금액과 팁의 회귀 분석
sns.lmplot(data=df_tips, x='total_bill', y='tip', hue='smoker')
```

#### 4.2.7. 히트맵(heatmap)

데이터의 상관계수나 빈도수 등을 수치에 따라 색상의 농도로 나타내는 그래프다. 주로 여러 변수 간의 상관관계를 파악하거나, 시간대별 데이터의 밀도를 확인할 때 사용한다.

- 특징: 행렬 데이터(Matrix)를 입력으로 받으며, 상관관계가 높을수록 짙거나 밝은색으로 표현되어 직관적인 해석이 가능하다.

- 주요 파라미터:

  - `annot=True`: 각 칸 안에 수치를 표시한다.

  - `cmap`: 색상 지도(Color Map)를 설정한다. (예: 'coolwarm', 'RdBu', 'viridis')

  - `fmt=".2f"`: 표시될 수치의 소수점 자릿수를 지정한다.

```python
# 코드 16. 상관계수 행렬 계산 (수치형 데이터만 선택)
# 상관 계수 구하기
corr = df_tips.corr(numeric_only=True)

# 히트맵 생성
sns.heatmap(corr, annot=True, fmt=".2f", cmap='coolwarm')
```

## 5. Matplotlib으로 보정


시각화 라이브러리의 뿌리인 Matplotlib을 직접 제어하면 여러 개의 그래프를 한눈에 보이게 배치하거나, 기본 설정으로는 부족한 세부 디자인을 수정할 수 있다.

```python
# 코드 17. 예제 데이터 로드
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

pd.options.display.max_rows = 6  # 6행까지만 출력

# 실습용 데이터 로드

cols = ['time', 'day', 'smoker', 'sex']
df_tips = sns.load_dataset('tips').astype({k: 'str' for k in cols})
df_iris = sns.load_dataset('iris')
```

### 5.1. Subplot 사용: 여러 그래프 한 번에 배치하기

`plt.subplots`를 사용하면 도화지를 여러 칸으로 나누어 다양한 관점의 그래프를 하나의 이미지로 만들 수 있다.

```python
# 코드 18. subplot 실습

# 1. 1행 2열 구조의 도화지 생성
# fig: 전체 도화지 객체, ax: 각 칸(축)의 주소가 담긴 리스트
fig, ax = plt.subplots(1, 2, figsize=(15, 6))

# 2. 첫 번째 칸(ax[0])에 Pandas의 bar 그래프 그리기
# 요일별 평균 요금을 집계하여 막대 그래프로 표현
df_tips.groupby('day')['total_bill'].mean().plot(kind='bar', ax=ax[0], color='lightcoral', rot=0)

# 3. 두 번째 칸(ax[1])에 Seaborn의 scatter 그래프 그리기
# 전체 금액과 팁의 상관관계를 요일별 색상으로 구분
sns.scatterplot(data=df_tips, x='total_bill', y='tip', hue='day', ax=ax[1])

# 4. 그래프 간 간격 최적화 및 출력
plt.tight_layout()
plt.show()
```

### 5.2. 다양한 보정: 그래프 디테일 다듬기

기본 그래프를 '보고서용'으로 만들기 위해 자주 사용하는 보정 기법들이다.

#### 5.2.1. 소제목 생성

여러 개의 그래프를 배치했을 때, 전체 제목(suptitle) 외에 각 그래프가 무엇을 의미하는지 개별적으로 설명해야 합니다. 이때 `ax.set_title()` 함수를 사용합니다.

```python
# 코드 19. subplot에 소제목 생성

# 1. 1행 2열 구조의 도화지 생성
# fig: 전체 도화지 객체, ax: 각 칸(축)의 주소가 담긴 리스트
fig, ax = plt.subplots(1, 2, figsize=(15, 6))

# 2. 첫 번째 칸(ax[0])에 Pandas의 bar 그래프 그리기
# 요일별 평균 요금을 집계하여 막대 그래프로 표현
df_tips.groupby('day')['total_bill'].mean().plot(kind='bar', ax=ax[0], color='lightcoral', rot=0)
ax[0].set_title('Average Bill by Day (Pandas Plot)')

# 3. 두 번째 칸(ax[1])에 Seaborn의 scatter 그래프 그리기
# 전체 금액과 팁의 상관관계를 요일별 색상으로 구분
sns.scatterplot(data=df_tips, x='total_bill', y='tip', hue='day', ax=ax[1])
ax[1].set_title('Total Bill vs Tip (Seaborn Plot)')

# 4. 그래프 간 간격 최적화 및 출력
plt.tight_layout()
plt.show()
```

#### 5.2.2. 제목과 축 라벨 설정
그래프가 무엇을 의미하는지 명확하게 전달하기 위해 필수적이다.

```python
# 코드 20. 제목과 축 라벨 설정

sns.boxplot(data=df_tips, x='day', y='total_bill')

# 제목 및 여백
plt.title('Daily Bill Analysis', fontsize=16, fontweight='bold', pad=20)
plt.xlabel('Day of the Week', fontsize=12) # X축 라벨
plt.ylabel('Bill Amount ($)', fontsize=12) # Y축 라벨
```

#### 5.2.3. 범례 수정

데이터를 가리지 않도록 범례(Legend) 위치를 옮기거나 이름을 가독성 있게 바꾼다.

```python
# 코드 21. 범례 수정
sns.scatterplot(data=df_tips, x='total_bill', y='tip', hue='time')
plt.legend(title='Meal Time', loc='upper left', frameon=True) # 왼쪽 상단 배치
plt.show()
```

#### 5.2.4. 테두리 및 그리드

- 테두리(Spines) 및 그리드 추가 또는 제거: 불필요한 요소를 제거하여 데이터 자체에 집중하게 만든다. (Clean Visualization)

```python
# 코드 22. 테두리 및 그리드 추가
ax = sns.lineplot(data=df_tips, x=df_tips.index, y='total_bill')

# 상단과 오른쪽 테두리 제거
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)

plt.grid(axis='y', linestyle='--', alpha=0.5) # Y축에만 점선 그리드 추가
plt.show()
```
