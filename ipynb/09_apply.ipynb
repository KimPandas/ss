````python
# ì½”ë“œ 9-32. ìœ„ë„ì™€ ê²½ë„ê°€ ì•Œë ¤ì§„ ë‘ ì§€ì ì˜ ì§ì„ ê±°ë¦¬ êµ¬í•˜ê¸°
import haversine
seoul = [37.51, 127.05] # ìœ„ë„, ê²½ë„
pusan = [35.12, 129.08]
haversine.haversine(seoul, pusan, unit='km')
````

````python
# ì½”ë“œ 9-33. ê° ë„ì‹œì™€ ì„œìš¸ì˜ ì§ì„ ê±°ë¦¬ë¥¼ êµ¬í•  ì‹¤ìŠµ ì˜ˆì œ ì½”ë“œ
data2 = {'ë„ì‹œ': ['ì„œìš¸', 'ë¶€ì‚°', 'ê´‘ì£¼', 'ëŒ€ì „', 'ê°•ë¦‰', 'ëŒ€êµ¬'],
         'ìœ„ë„': [37.51, 35.12, 35.13, 36.35, 37.75, 35.87],
         'ê²½ë„': [127.05, 129.08, 126.9, 127.42, 128.88, 128.6]}
df2 = pd.DataFrame(data2)
df2
````

````python
# ì½”ë“œ 9-34. lambda í•¨ìˆ˜ ìƒì„±ì„ ìœ„í•´ df2ì˜ ë‘ ë²ˆì§¸ í–‰ì„ ë³€ìˆ˜ xë¡œ ì§€ì •í•˜ê¸°
x = df2.iloc[1]
x
````

````python
# ì½”ë“œ 9-35. ë³€ìˆ˜ xì™€ ë³€ìˆ˜ seoul, í•¨ìˆ˜ haversineìœ¼ë¡œ ì„œìš¸, ë¶€ì‚° ê±°ë¦¬ êµ¬í•˜ê¸°
haversine.haversine(seoul, [x['ìœ„ë„'], x['ê²½ë„']], unit='km')
````

````python
# ì½”ë“œ 9-36. df2ì—ì„œ ê° ë„ì‹œì™€ ì„œìš¸ê³¼ì˜ ê±°ë¦¬ êµ¬í•˜ê¸°
df2['ê±°ë¦¬(ì„œìš¸)'] = df2.apply(
    lambda x: haversine.haversine(seoul, [x['ìœ„ë„'], x['ê²½ë„']], unit='km'),
    axis=1
)
df2
````
